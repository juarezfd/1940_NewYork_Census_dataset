[How to Create a New Column Based on a Condition in Pandas](https://www.statology.org/create-column-based-on-condition-pandas/)

[Pandas Check Column Contains a Value in DataFrame](https://sparkbyexamples.com/pandas/pandas-check-column-contains-a-value-in-dataframe/#:~:text=Pandas%20%2F%20Python%20You%20can%20check%20if%20a,a%20column%20contains%20a%20particular%20value%20with%20examples.)

```python
#  Create a New Column with Multiple Values using a funciton
def male_abscent(row):
    if row['hhtype'] == 'female householder, living alone':
        val = True
    elif row['hhtype'] == 'female householder, living alone':
        val = True
    elif row['hhtype'] == 'female householder, not living alone':
        val = True
    else:
        val = False
    return val
```

```python
df['male_abscent'] = df.apply(male_abscent, axis=1) # create new column 'male_abscent' using the above function
```

```python
df['labforce_bool'] = df['labforce'].str.contains('yes')
# new column having true/false
# if yes was in there then it's true else false
```

[Removing characters from a string in pandas](https://stackoverflow.com/questions/37919479/removing-characters-from-a-string-in-pandas)

[Extract only numbers and only string from pandas dataframe](https://stackoverflow.com/questions/60295698/extract-only-numbers-and-only-string-from-pandas-dataframe)

[How to extract specific content in a pandas dataframe with a regex?](https://stackoverflow.com/questions/36028932/how-to-extract-specific-content-in-a-pandas-dataframe-with-a-regex)

```python
# creating a new column with information from another
# extracting only the numerical values 
df['nchild_int'] = df['nchild'].str.extract(r'(\d+[.\d]*)') 
df['famsize_int'] = df['famsize'].str.extract(r'(\d+[.\d]*)')
```

```python
df['nchild_int'].unique() 
```

```python
df['nchild_int'].sample() # returns one row in the series/column
```

```python
df.info() # information like datatype on the entire dataframe/table
```

[Change the data type of a column or a Pandas Series](https://www.geeksforgeeks.org/change-the-data-type-of-a-column-or-a-pandas-series/)

```python
df[['nchild_int']] = df[['nchild_int']].apply(pd.to_numeric) 
df[['famsize_int']] = df[['famsize_int']].apply(pd.to_numeric) 
```

```python
print(df['nchild_int'].dtype, df['nchild_int'].dtype) # checking a columns datatype
```
